1. Desarrolle un programa que permita realizar la suma/resta de matrices. El programa debe validar las condicielementoss necesarias para ejecutar la operación.

```python
def matrices():
    matrices_lista = []
    for i in range(2):
        filas = int(input(f"¿Cuantas filas quiere que tenga la matriz {i+1}?: "))
        matriz = []

        for j in range(filas): 
            e = input(f"Ingrese fila {j+1} (Separe los numeros con -): ")
            sp_e = e.split("-") # Separa los numeros ingresados
            
            elementos = []
            for k in sp_e: # Como split es un metodo de str toca pasarlos a int
                elementos.append(int(k))
            matriz.append(elementos)
        matrices_lista.append(matriz)

    return matrices_lista

def verificacion(matriz_a:list, matriz_b:list):
    # Las filas de cada matriz tiene el mismo tamaño
    columnas_a = len(matriz_a[0])
    for i in matriz_a: 
        if len(i) == columnas_a:
            pass
        else:
            return "Error"
    columnas_b = len(matriz_b[0])
    for i in matriz_b:
        if len(i) == columnas_b:
            pass
        else:
            return "Error"
    
    if len(matriz_a) == len(matriz_b): # Mismo número de filas
        if columnas_a == columnas_b: # Mismo número de columnas
            pass
    else:
        return "Error"
        
def calculo_suma(matriz_a:list, matriz_b:list):
    matriz_suma = []
    for i in range(len(matriz_a)):
        filas = []
        for j in range(len(matriz_a[i])):
            elementos = matriz_a[i][j] + matriz_b[i][j] # Suma los elementos respetivos de cada matriz
            filas.append(elementos)
        matriz_suma.append(filas)

    return matriz_suma

def calculo_resta(matriz_a:list, matriz_b:list):
    matriz_resta = []
    for i in range(len(matriz_a)):
        filas = []
        for j in range(len(matriz_a[i])):
            elementos = matriz_a[i][j] - matriz_b[i][j] # Resta los elementos respetivos de cada matriz
            filas.append(elementos)
        matriz_resta.append(filas)

    return matriz_resta

def suma_resta(sumaa, restaa): # El usuario escoge si sumar o restar las matrices
    while True:
        usuario = input("¿Quiere sumar o restar las matrices?: ")
        if usuario == "Suma":
            return sumaa
        elif usuario == "Resta":
            return restaa
        else:
            print("Opción no valida")

if __name__ == "__main__":
    matriz_a, matriz_b = matrices() # Separa la matrices_lista en 2
    condiciones = verificacion(matriz_a, matriz_b)

    if condiciones == "Error": # Verificacion de las condiciones de suma o resta
        print("No cumple con las condicones, ingrese nuevamente las matrices.")
    else:
        sumaa = calculo_suma(matriz_a, matriz_b)
        restaa = calculo_resta(matriz_a, matriz_b)
        res = suma_resta(sumaa, restaa)
        print("Matriz A:")
        for i in range(len(matriz_a)): # Imprimir en forma de matriz la lista de listas
            print(matriz_a[i])
        print("\nMatriz B:")
        for i in range(len(matriz_b)):
            print(matriz_b[i])
        print("\nA + B:")
        for i in range(len(res)):
            print(res[i])
```

2. Desarrolle un programa que permita realizar el producto de matrices. El programa debe validar las condiciones necesarias para ejecutar la operación.

```python
def matrices():
    matrices_lista = []
    for i in range(2):
        filas = int(input(f"¿Cuantas filas quiere que tenga la matriz {i+1}?: "))
        matriz = []

        for j in range(filas): 
            e = input(f"Ingrese fila {j+1} (Separe los numeros con -): ")
            sp_e = e.split("-") # Separa los numeros ingresados
            
            elementos = []
            for k in sp_e: # Como split es un metodo de str toca pasarlos a int
                elementos.append(int(k))
            matriz.append(elementos)
        matrices_lista.append(matriz)

    return matrices_lista

def verificacion(matriz_a:list, matriz_b:list):
    # Las filas de cada matriz tiene el mismo tamaño
    columnas_a = len(matriz_a[0])
    for i in matriz_a: 
        if len(i) == columnas_a:
            pass
        else:
            return "Error"
    columnas_b = len(matriz_b[0])
    for i in matriz_b:
        if len(i) == columnas_b:
            pass
        else:
            return "Error"
    
    if len(matriz_a[0]) == len(matriz_b): # Mismo número de columnas de a que filas de b
        pass
    else:
        return "Error"
        
def calculo_multiplicaion(matriz_a:list, matriz_b:list):
    matriz_multi = []
    for i in range(len(matriz_a)): # Recorre filas de a
        filas = []
        for j in range(len(matriz_b[0])): # Recorre columnas de b
            suma : int = 0 # Inicializar suma
            for k in range(len(matriz_b)): # Itera para la sumatoria de la multiplicacion
                suma += matriz_a[i][k] * matriz_b[k][j]
            filas.append(suma)
        matriz_multi.append(filas)
    return matriz_multi

if __name__ == "__main__":
    matriz_a, matriz_b = matrices() # Separa la matrices_lista en 2
    condiciones = verificacion(matriz_a, matriz_b)

    if condiciones == "Error": # Verificacion de las condiciones de suma o resta
        print("No cumple con las condicones, ingrese nuevamente las matrices.")
    else:
        multiplicaion = calculo_multiplicaion(matriz_a, matriz_b)
        print("Matriz A:")
        for i in range(len(matriz_a)): # Imprimir en forma de matriz la lista de listas
            print(matriz_a[i])
        print("\nMatriz B:")
        for i in range(len(matriz_b)):
            print(matriz_b[i])
        print("\nA * B:")
        for i in range(len(multiplicaion)):
            print(multiplicaion[i])
```

3. Desarrolle un programa que permita obtener la matriz transpuesta de una matriz ingresada. El programa debe validar las condiciones necesarias para ejecutar la operación.

```python
def matriz():
    filas = int(input(f"¿Cuantas filas quiere que tenga la matriz?: "))
    lista = []

    for j in range(filas): 
        e = input(f"Ingrese fila {j+1} (Separe los numeros con -): ")
        sp_e = e.split("-") # Separa los numeros ingresados
        
        elementos = []
        for k in sp_e: # Como split es un metodo de str toca pasarlos a int
            elementos.append(int(k))
        lista.append(elementos)

    return lista

def verificacion(matriz_original:list):
    # Las filas de cada matriz tiene el mismo tamaño
    columnas = len(matriz_original[0])
    for i in matriz_original: 
        if len(i) == columnas:
            pass
        else:
            return "Error"

def transpuesta(matriz_original:list):
    lista_t = []
    for i in range(len(matriz_original[0])): # Colunmas de la original
        elementos = []
        for j in range(len(matriz_original)): # Filas de la original
            elementos.append(matriz_original[j][i]) # Intercambio de filas por columnas
        lista_t.append(elementos)
    return lista_t

if __name__ == "__main__":
    matriz_original = matriz()
    condiciones = verificacion(matriz_original)
    
    if condiciones == "Error": 
        print("No cumple con las condicones, ingrese nuevamente la matriz.")
    else:
        matriz_transpuesta = transpuesta(matriz_original)
        print("Martiz original:")
        for i in range(len(matriz_original)): # Imprimir en forma de matriz la lista de listas
            print(matriz_original[i])
        print("\nMatriz transpuesta:")
        for i in range(len(matriz_transpuesta)):
            print(matriz_transpuesta[i])
```

4. Desarrollar un programa que sume los elementos de una columna dada de una matriz.

```python
def matriz():
    filas = int(input(f"¿Cuantas filas quiere que tenga la matriz?: "))
    lista = []

    for j in range(filas): 
        e = input(f"Ingrese fila {j+1} (Separe los numeros con -): ")
        sp_e = e.split("-") # Separa los numeros ingresados
        
        elementos = []
        for k in sp_e: # Como split es un metodo de str toca pasarlos a int
            elementos.append(int(k))
        lista.append(elementos)

    return lista

def verificacion(matriz_a:list):
    # Las filas de la matriz tiene el mismo tamaño
    columnas = len(matriz_a[0])
    for i in matriz_a: 
        if len(i) == columnas:
            pass
        else:
            return "Error"

def suma_columna(matriz_a:list):
    suma : int = 0 # Inicializar suma de columna
    columna = int(input("¿Que columna quiere sumar: "))
    columna -= 1 # La primera columna tiene indice 0
    if columna <= len(matriz_a[0]):
        for i in range(len(matriz_a)):
            suma += matriz_a[i][columna]
        return columna, suma
    else:
        print("Error, ingrese nuevamente la fila que desea sumar")


if __name__ == "__main__":
    matriz_a = matriz()
    condiciones = verificacion(matriz_a)
    
    if condiciones == "Error":
        print("No cumple con las condicones, ingrese nuevamente la matriz.")
    else:
        j_indice, resultado = suma_columna(matriz_a)
        print("Martiz:")
        for i in range(len(matriz_a)): # Imprimir en forma de matriz la lista de listas
            print(matriz_a[i])
        print()
        print(f"Suma columna {j_indice+1} → {resultado}")
```

5. Desarrollar un programa que sume los elementos de una fila dada de una matriz.

```python
def matriz():
    filas = int(input(f"¿Cuantas filas quiere que tenga la matriz?: "))
    lista = []

    for j in range(filas): 
        e = input(f"Ingrese fila {j+1} (Separe los numeros con -): ")
        sp_e = e.split("-") # Separa los numeros ingresados
        
        elementos = []
        for k in sp_e: # Como split es un metodo de str toca pasarlos a int
            elementos.append(int(k))
        lista.append(elementos)

    return filas, lista

def verificacion(matriz_a:list):
    # Las filas de la matriz tiene el mismo tamaño
    columnas = len(matriz_a[0])
    for i in matriz_a: 
        if len(i) == columnas:
            pass
        else:
            return "Error"

def suma_fila(filas:int, matriz_a:list):
    suma : int = 0 # Inicializar suma de fila
    fila = int(input("¿Que fila quiere sumar: "))
    fila -= 1 # La primera fila tiene indice 0

    while True: # Evitar que se trate de sumar una fila que no exista
        if fila <= filas:
            for i in matriz_a[fila]:
                suma += i
            return fila, suma
        else:
            print("Error, ingrese nuevamente la fila que desea sumar")
            return None

if __name__ == "__main__":
    filas, matriz_a = matriz()
    condiciones = verificacion(matriz_a)
    
    if condiciones == "Error":
        print("No cumple con las condicones, ingrese nuevamente la matriz.")
    else:
        indice, resultado = suma_fila(filas, matriz_a)
        print("Martiz:")
        for i in range(len(matriz_a)): # Imprimir en forma de matriz la lista de listas
            print(matriz_a[i])
        print()
        print(f"Suma fila {indice+1}: {matriz_a[indice]} → {resultado}")
```
